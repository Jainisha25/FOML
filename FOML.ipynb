```python
#Importing Libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Load the Dataset:

```python
housing = pd.read_csv('/content/FOMLHousing.csv')
```

```python
housing.head()
```

##Measure of Central Tendency:

```python
#Mean:
Mean = housing['area'].mean()
Mean
```

```python
#Median:
Median = housing['area'].median()
Median
```

```python
#Mode:
Mode = housing['area'].mode()
Mode
```

##Histogram:

```python
plt.hist(housing['area'],bins = 10)
```

##Boxplot:

```python
sns.boxplot(housing['area'])
```

##Scatter Plot:

```python
sns.scatterplot(data= housing, x = 'area', y ='price')
```

##Covariance and Correlation:

```python
#Covariance:
Cov = np.stack((housing['area'], housing['price']), axis =1)
print(np.cov(Cov))
```

```python
#Correlation:
Cor = housing.corr()
Cor
```
